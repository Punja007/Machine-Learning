```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, SnowballStemmer
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
df
```

```python
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],inplace=True)
```

```python
df
```

```python
df['v2']
```

```python
new = []
lmtz = WordNetLemmatizer()

for text in df['v2'].astype(str):  
    temp = re.sub(r'[^a-zA-Z]', ' ', text.lower())  # Remove non-alphabetic characters
    words = temp.split()  # Split into words
    words = [word for word in words if word not in stopwords.words('english')]
    lemmatized_words = [lmtz.lemmatize(word, pos='v') for word in words]  # Lemmatize all words as verbs
    new.append(" ".join(lemmatized_words))  # Join words back into a sentence

print(new[4])  # Print first processed text
```

```python
print(df['v2'][4])
```

```python
df['new_text'] = new
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x = df['new_text']
y = df['v1']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
x_train = tfidf.fit_transform(x_train).toarray()
x_test = tfidf.fit_transform(x_test).toarray()
np.set_printoptions(threshold=np.inf)
```
