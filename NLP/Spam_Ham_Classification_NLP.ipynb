```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, SnowballStemmer
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
df
```

```python
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],inplace=True)
```

```python
df
```

```python
df['v2']
```

```python
new = []
lmtz = WordNetLemmatizer()

for text in df['v2'].astype(str):  
    temp = re.sub(r'[^a-zA-Z]', ' ', text.lower())  # Remove non-alphabetic characters
    words = temp.split()  # Split into words
    lemmatized_words = [lmtz.lemmatize(word, pos='v') for word in words]  # Lemmatize all words as verbs
    new.append(" ".join(lemmatized_words))  # Join words back into a sentence

print(new[4])  # Print first processed text
```

```python
import xgboost
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
x = df1
y = df['v1']
```

```python
y = y.map({'ham': 1, 'spam': 0})
print(y)
```

```python
from imblearn.under_sampling import RandomUnderSampler

X_resampled, y_resampled = RandomUnderSampler().fit_resample(x, y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_resampled, y_resampled,test_size=0.2,random_state=42)
```

```python
y_train
```

```python
xgb = XGBClassifier()
xgb.fit(x_train,y_train)

train_pred = xgb.predict(x_train)
test_pred = xgb.predict(x_test)
```

```python
print(classification_report(y_train,train_pred))
print(classification_report(y_test,test_pred))
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)

train_pred = dtree.predict(x_train)
test_pred = dtree.predict(x_test)
```

```python
print(classification_report(y_train,train_pred))
print(classification_report(y_test,test_pred))
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)

train_pred = lr.predict(x_train)
test_pred = lr.predict(x_test)
```

```python
print(classification_report(y_train,train_pred))
print(classification_report(y_test,test_pred))
```
