```python
import pandas as pb
import seaborn as sns
```

```python
df = pb.read_csv('Titanic.csv')
df
```

```python
df.isna().sum()
```

```python
df.pop('Cabin')
df.pop('Name')
```

```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df.corr()
```

```python
sns.heatmap(numeric_df.corr(), annot=True, cmap='RdYlGn')
```

```python
sns.boxplot(data=df, y='Age')
```

```python
sns.jointplot(data=df, x="Age", y="Fare", kind="reg")
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="Sex")
```

```python
sns.displot(df.Fare, kde=True, bins=10)
```

Barplot is same as Count plot but we have to specify both X and Y axis

```python
sns.barplot(data=df, x="Sex", y="Pclass")
```

```python
sns.boxplot(x='Sex',y='Age',data=df,palette='rainbow')
```

```python
sns.boxplot(x='Embarked',y='Age',data=df,palette='rainbow')
```

Violinplot is same as boxplot just the diff is that it shows probablity distribution instead of counts

```python
sns.violinplot(x='Sex', y='Age', data=df)
```

```python
sns.violinplot( x='Embarked',y='Age', data=df)
```
